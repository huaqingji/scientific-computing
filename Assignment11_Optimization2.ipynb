(a) Graphically

```python
# 1. define and plot for constraint functions, for feasible solution space
# -----------------------------------------------------------------------
def constraint_f1(x):   
    result = -2.5 * x + 20
    return result

def constraint_f2(x):    
    result = -x + 10
    return result

def constraint_f3(x):    
    result = -0.5 * x + 8
    return result

x = np.linspace(-1, 10, 20)
y1 = constraint_f1(x)
y2 = constraint_f2(x)
y3 = constraint_f3(x)

plt.figure(figsize = (8, 5))
plt.plot(x, y1, 'b', label='constraint: 5x + 2y <= 40') # constraint 1
plt.plot(x, y2, 'g', label='constraint: 6x + 6y <= 60') # constraint 2
plt.plot(x, y3, 'm', label='constraint: 2x + 4y <= 32') # constraint 3
plt.axvline(0, c='c', label='constraint: x >= 0') # constraint 4
plt.axhline(0, c='y', label='constraint: y >= 0') # constraint 5
plt.title('Linear Programming')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.legend(loc='upper right')


# 2. plot all extreme points (where two constraints meet) in the feasible solution space
# --------------------------------------------------------------------------------------------
plt.plot(0, 0, 'ok', markersize=8) # extreme point 1
plt.text(0, 0, '(0, 0)')
plt.plot(0, 8, 'ok', markersize=8) # extreme point 2
plt.text(0, 8, '(0, 8)')
plt.plot(4, 6, 'ok', markersize=12) # extreme point 3
plt.text(4, 6, '(4, 6)')
plt.plot(8, 0, 'ok', markersize=8) # extreme point 4
plt.text(8, 0, '(8, 0)')
plt.plot(20/3, 10/3, 'ok', markersize=8) # extreme point 5
plt.text(20/3, 10/3, '(20/3, 10/3)')


# 3. define and plot the objective function (get the highest value from the extreme point)
# -----------------------------------------------------------------------------------------
def objective_f(x, z):   
    result = -6/8 * x + 1/8 * z
    return result

of1 = objective_f(x, 0)
plt.plot(x, of1, 'r--', label='objective: z = 0') # objective function for extreme point (0, 0)
of2 = objective_f(x, 64)
plt.plot(x, of2, 'r--', label='objective: z = 64') # objective function for extreme point (0, 8)
of3 = objective_f(x, 72)
plt.plot(x, of3, 'r--', label='objective: z = 72')  # objective function for extreme point (4, 6)
plt.legend(loc='upper right')
```

Result: so the extreme point (4, 6) should be the proper (x, y) with a maximum value 72

(c) Software Package

```python
# use the Scipy package in Python
# note: to minimize
from scipy.optimize import linprog

c = [-6, -8] # coeffcients of the objective function, make it negative!
A = [[5, 2], [6, 6], [2, 4]] # coefficients of the inequality constraints
b = [40, 60, 32] # right-hand side vector of the inequality constraints
x_bounds = (0, None) # lower and upper bounds
y_bounds = (0, None) # lower and upper bounds
linprog(c, A_ub=A, b_ub=b, bounds=[x_bounds, y_bounds])
```
