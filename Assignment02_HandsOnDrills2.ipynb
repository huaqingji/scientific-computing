### Probelm 2.18

Piecewise functions are sometimes useful when the relation-ship between a dependent and an independent variable cannot be adequately  represented  by  a  single  equation. For example, the velocity of a rocket might be described by

$$
υ(t) = \left\{
    \begin{array}\\
        11t^2 − 5t & \mbox 0 ≤  t ≤ 10\\
        1100 − 5t & \mbox 10 ≤ t ≤ 20\\
        50t + 2(t − 20)^2  & \mbox 20 ≤ t ≤ 30\\
        1520e^{−0.2(t − 30)}  & \mbox t > 30\\
        0  & \mbox otherwise
    \end{array}
\right.
$$

Develop a well-structured function to compute v as a function of t. 
Then use this function to generate a table of v versus t for t = −5 to 
50 at increments of 0.5.

### Answer
1. Develop a well-structured function to compute v as a function of t

```python
def velocity(t):
    """
    This function calculates the velocity of a rocket in based on t.
    """
    
    if t >= 0 and t <= 10:
        return 11*(t**2)-5*t
    elif t > 10 and t <= 20:
        return 1100-5*t
    elif t > 20 and t <= 30:
        return 50*t+2*(t-20)**2
    elif t > 30:
        return 1520*np.exp(-0.2*(t-30))
    else:
        return 0
```

2. Then use this function to generate a table of v versus t for t = −5 to 50 at increments of 0.5

```python
results = [[t, round(velocity(t), 2)] for t in np.arange(-5, 50.5, 0.5)]
df = pd.DataFrame(results, columns=['t', 'v(t)'])
df
```

```python
# plot the results
plt.figure(figsize = (8, 6))
plt.plot(df['t'], df['v(t)'], 'bo')
plt.xlim(-5, 50)
plt.ylim(0, 2000)
plt.title('The velocity versus t')
plt.xlabel('t')
plt.ylabel('v(t)')
plt.grid()
```
